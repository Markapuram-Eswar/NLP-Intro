## 1. Advanced Tokenization Examples


```python
from nltk.tokenize import word_tokenize, sent_tokenize, TweetTokenizer, MWETokenizer

# Sample text for demonstration
sample_text = """
Natural Language Processing (NLP) is a fascinating field of artificial intelligence! 
It helps computers understand, interpret, and manipulate human language. 
Companies like Google, Microsoft, and OpenAI use NLP for various applications.
#NLP #AI #MachineLearning @OpenAI
"""

print("ðŸ“ Sample Text:")
print(sample_text)
print("\n" + "="*60)

# 1. Sentence Tokenization
print("ðŸ”¤ SENTENCE TOKENIZATION:")
sentences = sent_tokenize(sample_text)
for i, sentence in enumerate(sentences, 1):
    print(f"{i}. {sentence.strip()}")

print("\n" + "="*60)

# 2. Word Tokenization
print("ðŸ”¤ WORD TOKENIZATION:")
words = word_tokenize(sample_text)
print(f"Total words: {len(words)}")
print(f"Words: {words}")

print("\n" + "="*60)

# 3. Tweet Tokenization (handles hashtags, mentions, emoticons)
print("ðŸ¦ TWEET TOKENIZATION:")
tweet_tokenizer = TweetTokenizer()
tweet_tokens = tweet_tokenizer.tokenize(sample_text)
print(f"Tweet tokens: {tweet_tokens}")

print("\n" + "="*60)

# 4. Multi-Word Expression Tokenization
print("ðŸ”— MULTI-WORD EXPRESSION TOKENIZATION:")
mwe_tokenizer = MWETokenizer([('Natural', 'Language', 'Processing'), ('artificial', 'intelligence')])
mwe_tokens = mwe_tokenizer.tokenize(word_tokenize(sample_text))
print(f"MWE tokens: {mwe_tokens}")
```

## 2. Text Preprocessing Pipeline


```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
import re

class TextPreprocessor:
    def __init__(self):
        self.stemmer = PorterStemmer()
        self.lemmatizer = WordNetLemmatizer()
        self.stop_words = set(stopwords.words('english'))
        
    def clean_text(self, text):
        """Remove special characters, numbers, and extra whitespace"""
        # Convert to lowercase
        text = text.lower()
        
        # Remove URLs
        text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
        
        # Remove email addresses
        text = re.sub(r'\S+@\S+', '', text)
        
        # Remove special characters and digits
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        
        # Remove extra whitespace
        text = re.sub(r'\s+', ' ', text).strip()
        
        return text
    
    def remove_stopwords(self, tokens):
        """Remove stopwords from tokenized text"""
        return [token for token in tokens if token not in self.stop_words]
    
    def stem_tokens(self, tokens):
        """Apply stemming to tokens"""
        return [self.stemmer.stem(token) for token in tokens]
    
    def lemmatize_tokens(self, tokens):
        """Apply lemmatization to tokens"""
        return [self.lemmatizer.lemmatize(token) for token in tokens]
    
    def preprocess_pipeline(self, text, remove_stopwords=True, use_lemmatization=True):
        """Complete preprocessing pipeline"""
        # Clean text
        cleaned_text = self.clean_text(text)
        
        # Tokenize
        tokens = word_tokenize(cleaned_text)
        
        # Remove stopwords
        if remove_stopwords:
            tokens = self.remove_stopwords(tokens)
        
        # Apply lemmatization or stemming
        if use_lemmatization:
            tokens = self.lemmatize_tokens(tokens)
        else:
            tokens = self.stem_tokens(tokens)
        
        return tokens

# Example usage
preprocessor = TextPreprocessor()

# Sample messy text
messy_text = """
Hello! This is a SAMPLE text with numbers 123, URLs https://example.com, 
and EMAIL addresses like test@email.com. It has SPECIAL characters!!! 
and multiple    spaces. The text contains stopwords like 'the', 'and', 'is'.
"""

print("ðŸ§¹ ORIGINAL TEXT:")
print(messy_text)
print("\n" + "="*60)

print("ðŸ§¹ CLEANED TEXT:")
cleaned = preprocessor.clean_text(messy_text)
print(cleaned)
print("\n" + "="*60)

print("ðŸ”¤ TOKENIZED TEXT:")
tokens = word_tokenize(cleaned)
print(f"Tokens: {tokens}")
print("\n" + "="*60)

print("ðŸš« AFTER REMOVING STOPWORDS:")
tokens_no_stopwords = preprocessor.remove_stopwords(tokens)
print(f"Tokens: {tokens_no_stopwords}")
print("\n" + "="*60)

print("ðŸ“ LEMMATIZED TOKENS:")
lemmatized = preprocessor.lemmatize_tokens(tokens_no_stopwords)
print(f"Lemmatized: {lemmatized}")
print("\n" + "="*60)

print("ðŸ”„ COMPLETE PREPROCESSING PIPELINE:")
final_tokens = preprocessor.preprocess_pipeline(messy_text)
print(f"Final tokens: {final_tokens}")
```

## 3. Part-of-Speech (POS) Tagging


```python
from nltk.tag import pos_tag
from nltk.chunk import ne_chunk
from nltk.tokenize import word_tokenize

# Sample text for POS tagging
pos_text = "The quick brown fox jumps over the lazy dog. Natural Language Processing is amazing!"

print("ðŸ“ SAMPLE TEXT:")
print(pos_text)
print("\n" + "="*60)

# Tokenize the text
tokens = word_tokenize(pos_text)
print("ðŸ”¤ TOKENS:")
print(tokens)
print("\n" + "="*60)

# Perform POS tagging
pos_tags = pos_tag(tokens)
print("ðŸ·ï¸ POS TAGS:")
for word, tag in pos_tags:
    print(f"{word:15} -> {tag}")

print("\n" + "="*60)

# POS Tag explanations
pos_explanations = {
    'NN': 'Noun, singular or mass',
    'NNS': 'Noun, plural',
    'NNP': 'Proper noun, singular',
    'NNPS': 'Proper noun, plural',
    'VB': 'Verb, base form',
    'VBD': 'Verb, past tense',
    'VBG': 'Verb, gerund or present participle',
    'VBN': 'Verb, past participle',
    'VBP': 'Verb, non-3rd person singular present',
    'VBZ': 'Verb, 3rd person singular present',
    'JJ': 'Adjective',
    'JJR': 'Adjective, comparative',
    'JJS': 'Adjective, superlative',
    'RB': 'Adverb',
    'RBR': 'Adverb, comparative',
    'RBS': 'Adverb, superlative',
    'DT': 'Determiner',
    'IN': 'Preposition or subordinating conjunction',
    'CC': 'Coordinating conjunction',
    'PRP': 'Personal pronoun',
    'PRP$': 'Possessive pronoun'
}

print("ðŸ“š POS TAG EXPLANATIONS:")
for word, tag in pos_tags:
    if tag in pos_explanations:
        print(f"{word:15} -> {tag:5} ({pos_explanations[tag]})")
    else:
        print(f"{word:15} -> {tag:5}")

print("\n" + "="*60)

# Extract specific POS categories
nouns = [word for word, tag in pos_tags if tag.startswith('NN')]
verbs = [word for word, tag in pos_tags if tag.startswith('VB')]
adjectives = [word for word, tag in pos_tags if tag.startswith('JJ')]

print("ðŸ“Š EXTRACTED CATEGORIES:")
print(f"Nouns: {nouns}")
print(f"Verbs: {verbs}")
print(f"Adjectives: {adjectives}")
```

## 4. Named Entity Recognition (NER)


```python
from nltk.chunk import ne_chunk
from nltk.tag import pos_tag
from nltk.tokenize import word_tokenize

# Sample text with named entities
ner_text = """
Apple Inc. was founded by Steve Jobs in Cupertino, California in 1976. 
The company is headquartered at One Apple Park Way. 
Tim Cook is the current CEO of Apple.
Microsoft Corporation is based in Redmond, Washington.
"""

print("ðŸ“ SAMPLE TEXT:")
print(ner_text)
print("\n" + "="*60)

# Tokenize and POS tag
tokens = word_tokenize(ner_text)
pos_tags = pos_tag(tokens)

print("ðŸ”¤ TOKENS AND POS TAGS:")
for word, tag in pos_tags:
    print(f"{word:15} -> {tag}")

print("\n" + "="*60)

# Perform Named Entity Recognition
ner_tree = ne_chunk(pos_tags)

print("ðŸ·ï¸ NAMED ENTITY RECOGNITION:")
print(ner_tree)

print("\n" + "="*60)

# Extract named entities
print("ðŸ“Š EXTRACTED NAMED ENTITIES:")
def extract_entities(tree):
    entities = []
    for subtree in tree:
        if hasattr(subtree, 'label'):
            entity = ' '.join([token for token, pos in subtree.leaves()])
            entities.append((entity, subtree.label()))
        else:
            if len(subtree) == 2 and subtree[1] in ['NNP', 'NNPS']:
                entities.append((subtree[0], 'PERSON'))
    return entities

entities = extract_entities(ner_tree)
for entity, label in entities:
    print(f"{entity:25} -> {label}")

print("\n" + "="*60)

# Entity type explanations
entity_explanations = {
    'PERSON': 'People, including fictional',
    'ORGANIZATION': 'Companies, agencies, institutions',
    'GPE': 'Countries, cities, states',
    'FACILITY': 'Buildings, airports, highways',
    'LOCATION': 'Non-GPE locations, mountain ranges, bodies of water',
    'PRODUCT': 'Objects, vehicles, foods',
    'EVENT': 'Named hurricanes, battles, wars, sports events',
    'WORK_OF_ART': 'Titles of books, songs, etc.',
    'LAW': 'Named documents made into laws',
    'LANGUAGE': 'Any named language',
    'DATE': 'Absolute or relative dates or periods',
    'TIME': 'Times smaller than a day',
    'PERCENT': 'Percentage',
    'MONEY': 'Monetary values',
    'QUANTITY': 'Measurements',
    'ORDINAL': 'First, second, etc.',
    'CARDINAL': 'Numerals'
}

print("ðŸ“š ENTITY TYPE EXPLANATIONS:")
for entity, label in entities:
    if label in entity_explanations:
        print(f"{entity:25} -> {label:12} ({entity_explanations[label]})")
    else:
        print(f"{entity:25} -> {label}")
```

## 5. Sentiment Analysis


```python
from nltk.sentiment import SentimentIntensityAnalyzer
import pandas as pd

# Initialize the sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Sample texts for sentiment analysis
sample_texts = [
    "I love this product! It's absolutely amazing and works perfectly.",
    "This is terrible. I hate it and it doesn't work at all.",
    "The weather is okay today, nothing special.",
    "I'm so excited about the new movie! It looks fantastic!",
    "The service was poor and the staff was rude.",
    "This is a neutral statement with no strong emotions.",
    "I'm extremely disappointed with the quality of this item.",
    "What a wonderful day! Everything is going great!"
]

print("ðŸ“ SENTIMENT ANALYSIS EXAMPLES:")
print("="*60)

# Analyze sentiment for each text
results = []
for i, text in enumerate(sample_texts, 1):
    scores = sia.polarity_scores(text)
    
    # Determine sentiment label
    if scores['compound'] >= 0.05:
        sentiment = 'Positive'
    elif scores['compound'] <= -0.05:
        sentiment = 'Negative'
    else:
        sentiment = 'Neutral'
    
    results.append({
        'Text': text,
        'Sentiment': sentiment,
        'Compound': scores['compound'],
        'Positive': scores['pos'],
        'Neutral': scores['neu'],
        'Negative': scores['neg']
    })
    
    print(f"\n{i}. {text}")
    print(f"   Sentiment: {sentiment}")
    print(f"   Scores: {scores}")

print("\n" + "="*60)

# Create a DataFrame for better visualization
df = pd.DataFrame(results)
print("ðŸ“Š SENTIMENT ANALYSIS RESULTS:")
print(df[['Text', 'Sentiment', 'Compound']].to_string(index=False))

print("\n" + "="*60)

# Sentiment distribution
sentiment_counts = df['Sentiment'].value_counts()
print("ðŸ“ˆ SENTIMENT DISTRIBUTION:")
for sentiment, count in sentiment_counts.items():
    print(f"{sentiment}: {count}")

print("\n" + "="*60)

# Detailed score analysis
print("ðŸ” DETAILED SCORE ANALYSIS:")
print("Compound Score Range: -1 (most negative) to +1 (most positive)")
print("Positive Score: Proportion of positive words")
print("Neutral Score: Proportion of neutral words")
print("Negative Score: Proportion of negative words")
print("\nScore Interpretation:")
print("- Compound >= 0.05: Positive sentiment")
print("- Compound <= -0.05: Negative sentiment")
print("- -0.05 < Compound < 0.05: Neutral sentiment")
```

## 6. Advanced NLP Features - TF-IDF and Text Similarity


```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Sample documents for TF-IDF analysis
documents = [
    "Natural Language Processing is a field of artificial intelligence.",
    "Machine learning algorithms can process and analyze text data.",
    "Deep learning models are used for various NLP tasks.",
    "Text preprocessing is an important step in NLP pipelines.",
    "Sentiment analysis helps understand emotions in text.",
    "Named entity recognition identifies important entities in text.",
    "Tokenization is the process of splitting text into words.",
    "Part-of-speech tagging assigns grammatical categories to words."
]

print("ðŸ“ SAMPLE DOCUMENTS:")
for i, doc in enumerate(documents, 1):
    print(f"{i}. {doc}")

print("\n" + "="*60)

# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(
    max_features=100,
    stop_words='english',
    ngram_range=(1, 2),  # Include both unigrams and bigrams
    min_df=1,
    max_df=0.8
)

# Fit and transform the documents
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

print("ðŸ”¢ TF-IDF MATRIX SHAPE:")
print(f"Shape: {tfidf_matrix.shape}")
print(f"Number of documents: {tfidf_matrix.shape[0]}")
print(f"Number of features: {tfidf_matrix.shape[1]}")

print("\n" + "="*60)

# Get feature names (vocabulary)
feature_names = tfidf_vectorizer.get_feature_names_out()
print("ðŸ“š TOP FEATURES (VOCABULARY):")
print(f"Total features: {len(feature_names)}")
print("First 20 features:", feature_names[:20])

print("\n" + "="*60)

# Calculate TF-IDF scores for each document
print("ðŸ“Š TF-IDF SCORES FOR EACH DOCUMENT:")
for i, doc in enumerate(documents):
    print(f"\nDocument {i+1}: {doc}")
    
    # Get TF-IDF scores for this document
    doc_scores = tfidf_matrix[i].toarray().flatten()
    
    # Get top 5 features for this document
    top_indices = np.argsort(doc_scores)[-5:][::-1]
    print("Top 5 features:")
    for idx in top_indices:
        if doc_scores[idx] > 0:
            print(f"  {feature_names[idx]:20} -> {doc_scores[idx]:.4f}")

print("\n" + "="*60)

# Calculate cosine similarity between documents
similarity_matrix = cosine_similarity(tfidf_matrix)

print("ðŸ”— COSINE SIMILARITY MATRIX:")
print("Similarity scores between documents:")
print("(Higher scores indicate more similar documents)")

# Display similarity matrix
print("\nDocument Similarity Matrix:")
print("     ", end="")
for i in range(len(documents)):
    print(f"Doc{i+1:2}", end="  ")
print()

for i in range(len(documents)):
    print(f"Doc{i+1:2}", end="  ")
    for j in range(len(documents)):
        print(f"{similarity_matrix[i][j]:.3f}", end="  ")
    print()

print("\n" + "="*60)

# Find most similar document pairs
print("ðŸŽ¯ MOST SIMILAR DOCUMENT PAIRS:")
similarities = []
for i in range(len(documents)):
    for j in range(i+1, len(documents)):
        similarities.append((i, j, similarity_matrix[i][j]))

# Sort by similarity score
similarities.sort(key=lambda x: x[2], reverse=True)

for i, (doc1_idx, doc2_idx, score) in enumerate(similarities[:3]):
    print(f"\n{i+1}. Similarity Score: {score:.4f}")
    print(f"   Document {doc1_idx+1}: {documents[doc1_idx]}")
    print(f"   Document {doc2_idx+1}: {documents[doc2_idx]}")

print("\n" + "="*60)

# Query-based similarity search
query = "machine learning and text analysis"
print(f"ðŸ” QUERY-BASED SIMILARITY SEARCH:")
print(f"Query: '{query}'")

# Transform query using the same vectorizer
query_vector = tfidf_vectorizer.transform([query])

# Calculate similarity with all documents
query_similarities = cosine_similarity(query_vector, tfidf_matrix).flatten()

# Get top 3 most similar documents
top_indices = np.argsort(query_similarities)[-3:][::-1]

print("\nTop 3 most similar documents:")
for i, idx in enumerate(top_indices):
    print(f"\n{i+1}. Similarity: {query_similarities[idx]:.4f}")
    print(f"   Document {idx+1}: {documents[idx]}")
```

```python
# Real-World NLP Applications and Use Cases

print("ðŸŒ REAL-WORLD NLP APPLICATIONS")
print("="*60)

# 1. Customer Review Analysis
print("1ï¸âƒ£ CUSTOMER REVIEW ANALYSIS")
print("Use Case: E-commerce platforms analyzing product reviews")

# Sample product reviews
product_reviews = [
    "This laptop is absolutely fantastic! The battery life is amazing and it runs smoothly.",
    "Terrible product. Broke after just one week. Waste of money.",
    "Decent quality for the price. Nothing extraordinary but gets the job done.",
    "Love this phone! The camera quality is outstanding and the design is beautiful.",
    "Poor customer service. The product arrived damaged and they refused to replace it.",
    "Excellent value for money. Highly recommend this product to everyone.",
    "Average product. Not bad but not great either. Could be better.",
    "Outstanding quality! This exceeded all my expectations. Will definitely buy again."
]

print("Sample Product Reviews:")
for i, review in enumerate(product_reviews, 1):
    print(f"{i}. {review}")

# Analyze sentiment for each review
sia = SentimentIntensityAnalyzer()
review_analysis = []

for review in product_reviews:
    scores = sia.polarity_scores(review)
    if scores['compound'] >= 0.05:
        sentiment = 'Positive'
    elif scores['compound'] <= -0.05:
        sentiment = 'Negative'
    else:
        sentiment = 'Neutral'
    
    review_analysis.append({
        'Review': review,
        'Sentiment': sentiment,
        'Compound_Score': scores['compound']
    })

print("\nðŸ“Š REVIEW SENTIMENT ANALYSIS:")
for i, analysis in enumerate(review_analysis, 1):
    print(f"{i}. {analysis['Sentiment']} (Score: {analysis['Compound_Score']:.3f}) - {analysis['Review'][:50]}...")

# Calculate overall sentiment
positive_count = sum(1 for a in review_analysis if a['Sentiment'] == 'Positive')
negative_count = sum(1 for a in review_analysis if a['Sentiment'] == 'Negative')
neutral_count = sum(1 for a in review_analysis if a['Sentiment'] == 'Neutral')

print(f"\nOverall Product Sentiment:")
print(f"Positive: {positive_count} ({positive_count/len(review_analysis)*100:.1f}%)")
print(f"Negative: {negative_count} ({negative_count/len(review_analysis)*100:.1f}%)")
print(f"Neutral: {neutral_count} ({neutral_count/len(review_analysis)*100:.1f}%)")

print("\n" + "="*60)

# 2. News Article Classification
print("2ï¸âƒ£ NEWS ARTICLE CLASSIFICATION")
print("Use Case: Automatically categorizing news articles by topic")

# Sample news articles
news_articles = [
    "Apple Inc. reported record quarterly earnings, with iPhone sales exceeding expectations.",
    "Scientists discover new exoplanet that could potentially support life.",
    "Local football team wins championship after thrilling overtime victory.",
    "New climate change legislation passes in Congress with bipartisan support.",
    "Tech startup raises $50 million in Series A funding round.",
    "Medical breakthrough: New treatment shows promise for cancer patients.",
    "Stock market reaches new all-time high amid economic optimism.",
    "SpaceX successfully launches satellite constellation into orbit."
]

# Define categories
categories = {
    'Technology': ['apple', 'iphone', 'tech', 'startup', 'funding', 'spacex', 'satellite'],
    'Science': ['scientists', 'exoplanet', 'life', 'medical', 'breakthrough', 'treatment', 'cancer'],
    'Sports': ['football', 'team', 'championship', 'victory'],
    'Politics': ['congress', 'legislation', 'bipartisan'],
    'Finance': ['earnings', 'sales', 'stock', 'market', 'economic']
}

def classify_article(article, categories):
    """Simple keyword-based classification"""
    article_lower = article.lower()
    scores = {}
    
    for category, keywords in categories.items():
        score = sum(1 for keyword in keywords if keyword in article_lower)
        scores[category] = score
    
    # Return category with highest score, or 'Other' if no matches
    if max(scores.values()) > 0:
        return max(scores, key=scores.get)
    else:
        return 'Other'

print("ðŸ“° NEWS ARTICLE CLASSIFICATION:")
for i, article in enumerate(news_articles, 1):
    category = classify_article(article, categories)
    print(f"{i}. [{category}] {article}")

print("\n" + "="*60)

# 3. Email Spam Detection
print("3ï¸âƒ£ EMAIL SPAM DETECTION")
print("Use Case: Filtering spam emails from legitimate ones")

# Sample emails
emails = [
    "Hi John, I hope you're doing well. Let's schedule a meeting for next week to discuss the project.",
    "URGENT! You've won $1,000,000! Click here to claim your prize now! Limited time offer!",
    "Thank you for your recent purchase. Your order has been shipped and will arrive in 3-5 business days.",
    "FREE MONEY! Make $5000 per week working from home! No experience required!",
    "The quarterly report is ready for review. Please find the attached document.",
    "CONGRATULATIONS! You've been selected for a special offer! Act now before it's too late!",
    "Meeting reminder: Team standup at 10 AM tomorrow in conference room A.",
    "Get rich quick! Invest in our amazing opportunity and double your money in 30 days!"
]

# Simple spam detection based on keywords
spam_keywords = ['urgent', 'free', 'money', 'congratulations', 'click here', 'limited time', 'act now', 'get rich', 'no experience', 'special offer']

def detect_spam(email, spam_keywords):
    """Simple spam detection based on keywords"""
    email_lower = email.lower()
    spam_score = sum(1 for keyword in spam_keywords if keyword in email_lower)
    
    # Also check for excessive punctuation and capitalization
    exclamation_count = email.count('!')
    caps_ratio = sum(1 for c in email if c.isupper()) / len(email) if email else 0
    
    if spam_score > 0 or exclamation_count > 2 or caps_ratio > 0.3:
        return 'Spam'
    else:
        return 'Legitimate'

print("ðŸ“§ EMAIL SPAM DETECTION:")
for i, email in enumerate(emails, 1):
    classification = detect_spam(email, spam_keywords)
    print(f"{i}. [{classification}] {email}")

print("\n" + "="*60)

# 4. Social Media Monitoring
print("4ï¸âƒ£ SOCIAL MEDIA MONITORING")
print("Use Case: Monitoring brand mentions and sentiment on social media")

# Sample social media posts
social_posts = [
    "Just tried the new iPhone 15 and it's absolutely incredible! #Apple #iPhone15",
    "Disappointed with the latest software update. Everything is so slow now. #TechProblems",
    "Love the new features in the latest app update! Great job team! #Innovation",
    "Customer service was terrible. Waited 2 hours on hold. #BadService",
    "Amazing product launch event! Can't wait to get my hands on this. #Excited",
    "The new design is okay, but I expected more for the price. #MixedFeelings",
    "Best purchase I've made this year! Highly recommend to everyone. #Satisfied",
    "Product arrived damaged and customer support is unresponsive. #Frustrated"
]

# Extract hashtags and analyze sentiment
def extract_hashtags(text):
    """Extract hashtags from text"""
    import re
    hashtags = re.findall(r'#\w+', text)
    return hashtags

def analyze_social_sentiment(posts):
    """Analyze sentiment of social media posts"""
    results = []
    for post in posts:
        hashtags = extract_hashtags(post)
        scores = sia.polarity_scores(post)
        
        if scores['compound'] >= 0.05:
            sentiment = 'Positive'
        elif scores['compound'] <= -0.05:
            sentiment = 'Negative'
        else:
            sentiment = 'Neutral'
        
        results.append({
            'Post': post,
            'Hashtags': hashtags,
            'Sentiment': sentiment,
            'Score': scores['compound']
        })
    
    return results

social_analysis = analyze_social_sentiment(social_posts)

print("ðŸ“± SOCIAL MEDIA SENTIMENT ANALYSIS:")
for i, analysis in enumerate(social_analysis, 1):
    print(f"{i}. [{analysis['Sentiment']}] {analysis['Post']}")
    if analysis['Hashtags']:
        print(f"   Hashtags: {', '.join(analysis['Hashtags'])}")

# Calculate overall brand sentiment
positive_posts = sum(1 for a in social_analysis if a['Sentiment'] == 'Positive')
negative_posts = sum(1 for a in social_analysis if a['Sentiment'] == 'Negative')
neutral_posts = sum(1 for a in social_analysis if a['Sentiment'] == 'Neutral')

print(f"\nðŸ“Š OVERALL BRAND SENTIMENT:")
print(f"Positive: {positive_posts} ({positive_posts/len(social_analysis)*100:.1f}%)")
print(f"Negative: {negative_posts} ({negative_posts/len(social_analysis)*100:.1f}%)")
print(f"Neutral: {neutral_posts} ({neutral_posts/len(social_analysis)*100:.1f}%)")

print("\n" + "="*60)

# 5. Chatbot Response Generation
print("5ï¸âƒ£ CHATBOT RESPONSE GENERATION")
print("Use Case: Creating intelligent responses for customer service chatbots")

# Sample customer queries
customer_queries = [
    "What are your business hours?",
    "How can I return a product?",
    "Do you offer international shipping?",
    "What payment methods do you accept?",
    "How do I track my order?",
    "Can I cancel my subscription?",
    "What is your refund policy?",
    "Do you have a mobile app?"
]

# Simple response generation based on keywords
def generate_response(query):
    """Generate appropriate response based on query keywords"""
    query_lower = query.lower()
    
    if any(word in query_lower for word in ['hours', 'open', 'close']):
        return "Our business hours are Monday-Friday 9 AM to 6 PM, and Saturday 10 AM to 4 PM."
    elif any(word in query_lower for word in ['return', 'refund']):
        return "You can return items within 30 days of purchase. Please contact our customer service for assistance."
    elif any(word in query_lower for word in ['shipping', 'delivery', 'international']):
        return "Yes, we offer international shipping to most countries. Shipping costs vary by location."
    elif any(word in query_lower for word in ['payment', 'pay', 'credit', 'debit']):
        return "We accept all major credit cards, PayPal, and bank transfers."
    elif any(word in query_lower for word in ['track', 'tracking', 'order']):
        return "You can track your order using the tracking number sent to your email."
    elif any(word in query_lower for word in ['cancel', 'subscription']):
        return "You can cancel your subscription anytime from your account settings."
    elif any(word in query_lower for word in ['app', 'mobile', 'application']):
        return "Yes, we have a mobile app available for both iOS and Android devices."
    else:
        return "I'd be happy to help you with that. Let me connect you with a customer service representative."

print("ðŸ¤– CHATBOT RESPONSE GENERATION:")
for i, query in enumerate(customer_queries, 1):
    response = generate_response(query)
    print(f"{i}. Customer: {query}")
    print(f"   Bot: {response}")
    print()

print("âœ… Real-world NLP applications demonstrated successfully!")
```
